```python
np.random.seed(42)

class Neuron:
    def __init__(self, n_inputs, n_neurons):
        self.weights = 0.1 * np.random.rand(n_neurons, n_inputs)
        self.bias_array = np.random.rand((1, n_neurons))

    def feedforward(self, inputs):
        output = np.dot(self.weights, inputs) + self.bias_array
        return output
    
    def activation_relu(self, output): # ReLU activation function
        return np.maximum(0, output)

    def activation_segmoid(self, output):
        return 1 / (1 + np.exp(-output))


    def forward(self, inputs):
        output = self.feedforward(inputs)
        activated_output = self.activation(output)
        return activated_output
```
